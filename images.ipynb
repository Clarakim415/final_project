```python
import os
import re
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service as ChromeService
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
import pandas as pd
import warnings
import time
import pickle

# 드라이버 불러오기
driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()))

# 피클 파일로 저장된 loaded_links 세트 로드
with open("links_559.pkl", "rb") as file:
    loaded_links = pickle.load(file)

# loaded_links 세트를 리스트로 변환
loaded_links = list(loaded_links)

# 이후에는 리스트로 슬라이싱 가능
for loaded_link in loaded_links[0:]:
    # 메인 페이지
    url = 'https://biz.koke.kr/products?filters=%7B%22categoryGroup%22%3A%22WholeBean%22%7D'
    driver.get(url)  # 웹 사이트로 이동

    # 이미지 URL과 관련된 텍스트(제품 이름)를 저장할 리스트
    image_info = []

    # loaded_links 리스트의 n번째 항목부터 시작
    for loaded_link in loaded_links[407:500]:
        driver.get(loaded_link)
        time.sleep(4)  # 페이지 로딩을 기다립니다.

        # 이미지를 찾기 위한 CSS 선택자
        group = driver.find_element(By.CSS_SELECTOR, 'div.flex.justify-center.items-center.relative.w-full.overflow-hidden.md\:rounded-lg.min-h-\[250px\].flex-shrink-0')

        # 이미지 URL 가져오기
        image_elements = group.find_elements(By.TAG_NAME, 'img')
        for image_element in image_elements:
            image_url = image_element.get_attribute('src')
            time.sleep(4)

        # 텍스트(제품 이름)를 찾기 위한 CSS 선택자
        pattern = r"https://biz.koke.kr/products/(\d+)"  # Regular expression pattern to match the product number
        match = re.search(pattern, loaded_link)
        if match:
            product_number = match.group(1)

        if image_url and product_number:
            image_info.append((image_url, product_number))
        time.sleep(4)
```

```python


output_directory = 'downloaded_images'
os.makedirs(output_directory, exist_ok=True)

for i, (image_url, product_number) in enumerate(image_info, start=1):
    response = requests.get(image_url)
    if response.status_code == 200:
        image_data = response.content
        image_extension = 'jpg'   # image_url.split('.')[-1],
        file_name = f"{product_number}.{image_extension}"  # 제품 번호를 파일 이름에 포함
        file_path = os.path.join(output_directory, file_name)
        with open(file_path,'wb') as image_file:
            image_file.write(image_data)
        print(f"Downloaded Image {i}: {file_name}")
    time.sleep(4)
```

```python
from PIL import Image
import os

input_directory = 'downloaded_images'
output_directory = 'converted_images'

os.makedirs(output_directory, exist_ok=True)

for filename in os.listdir(input_directory):
    if filename.endswith(('.png', '.jpeg', '.gif', '.bmp')):
        image_path = os.path.join(input_directory, filename)
        img = Image.open(image_path)

        # 이미지를 jpg로 저장
        jpg_filename = os.path.splitext(filename)[0] + '.jpg'
        jpg_path = os.path.join(output_directory, jpg_filename)
        img.convert('RGB').save(jpg_path)

        print(f"Converted {filename} to {jpg_filename}")

print("Conversion complete.")
```
