```python
import pandas as pd
```

```python
df = pd.read_csv('full_1.csv')
```

```python
import emoji

# 이모지를 처리하는 함수 정의
def process_country(value):
    if pd.notna(value):  # NaN 값이 아닌지 확인
        return emoji.replace_emoji(str(value), '')  # 문자열로 변환하고 이모지 제거
    else:
        return value  # NaN 값은 그대로 유지

# 'country' 열에 함수를 적용
df['country'] = df['country'].apply(process_country)

# 처리된 DataFrame 출력
df.head()
```

```python
df['country'] = df['country'].str.split(',')
df
```

```python
df['country'] = df['country'].fillna('s')
```

```python
country_list = ['에티오피아','브라질','콜롬비아','과테말라','인도','엘살바도르','예가체프','코스타리카','르완다','케냐','베트남','온두라스','탄자니아']
```

```python
def replace_country_names(countries):
    updated_countries = [next((country for country in country_list if country in entry), entry) for entry in countries]
    return updated_countries

# Apply the function to the 'country' column
df['country'] = df['country'].apply(replace_country_names)
```

```python
df['country'].tolist()
```

```python
# 문자열로 변환
df['country'] = df['country'].apply(lambda x: ', '.join(x))
```

```python
# s값 -> nan값으로 바꾸기
df['country'] = df['country'].apply(lambda x: x.replace('s', ''))
```
