```python
preprocessing = data.fillna(value = -1)     # data파일의 null -1로 바꿈
preprocessing
```

```python
for i in range(preprocessing.shape[0]):
    if preprocessing.loc[i, '성분'] != -1:         # -1로 바꾼 null값
        preprocessing['원재료 및 함량'][i] = preprocessing['성분'][i]
preprocessing
```

```python
# 필요없는 컬럼 드랍하기
preprocessed = preprocessing.drop(['보관방법', '제품명', '원재료', '용량', '포장재질', '제조원',
       '주소', '허가번호', '제조일', '권장기한', '생산자', '소재지', '제조연월일', '유통기한 또는 품질유지기한',
       '포장단위별 내용물의 용량(중량)', '원재료명 및 함량', '포장단위별 내용물의 용량(중량), 수량', '식품유형',
       '원재료명', '품질유지기한', '제조 및 판매원', '성분'], axis = 1)       # , inplace = True 
preprocessed
```

```python
import numpy as np
preprocessed = preprocessed.replace(-1, np.NaN)
```

```python
preprocessed[preprocessed['식품의 유형'].isnull()]
```

```python
# dataframe을 csv 파일로 저장하기
preprocessed.to_csv('./Required_information_preprocessed.csv', index = False, encoding='utf-8-sig')       # encoding='utf-8'이 한글 깨짐
```
